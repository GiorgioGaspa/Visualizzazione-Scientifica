```python
salary2023 = pd.read_csv('C:/Users/user/Desktop/ds_salaries2023.csv')
salary2024 = pd.read_csv('C:/Users/user/Desktop/Dataset salary 2024.csv')

avg_salary_2023 = salary2023['salary_in_usd'].mean()
avg_salary_2024 = salary2024['salary_in_usd'].mean()

years = ['2023', '2024']
avg_salaries = [avg_salary_2023, avg_salary_2024]

plt.figure(figsize=(10, 6))
plt.bar(years, avg_salaries)
plt.xlabel('Anno')
plt.ylabel('Stipendio Medio in USD')
plt.title('Comparazione Stipendio Medio: 2023 vs 2024')
plt.show()

```

```python
#https://www.kaggle.com/datasets/arnabchaki/data-science-salaries-2023
#https://www.kaggle.com/datasets/zeesolver/data-eng-salary-2024

avg_salary_by_experience_2023 = salary2023.groupby('experience_level')['salary_in_usd'].mean().reset_index()
avg_salary_by_experience_2024 = salary2024.groupby('experience_level')['salary_in_usd'].mean().reset_index()

avg_salary_by_experience = pd.merge(avg_salary_by_experience_2023, avg_salary_by_experience_2024, on='experience_level', suffixes=('_2023', '_2024'))

plt.figure(figsize=(10, 6))
bar_width = 0.35
index = range(len(avg_salary_by_experience))

plt.bar(index, avg_salary_by_experience['salary_in_usd_2023'], bar_width, label='2023')
plt.bar([i + bar_width for i in index], avg_salary_by_experience['salary_in_usd_2024'], bar_width, label='2024')

plt.xlabel('Livello di Esperienza')
plt.ylabel('Stipendio Medio in USD')
plt.title('Stipendi Medi per Livello di Esperienza: 2023 vs 2024')
plt.xticks([i + bar_width/2 for i in index], avg_salary_by_experience['experience_level'])
plt.legend()
plt.tight_layout()
plt.show()
```

```python
data_jobs_df = pd.read_csv('C:/Users/user/Desktop/jobs_in_data.csv')

average_salary_by_category = data_jobs_df.groupby('job_category')['salary_in_usd'].mean().reset_index()

average_salary_by_category.columns = ['Job Category', 'Average Salary in USD']


plt.figure(figsize=(12, 8))
plt.barh(average_salary_by_category['Job Category'], average_salary_by_category['Average Salary in USD'], color='#6495ED')
plt.xlabel('Stipendio Medio in USD')
plt.ylabel('Categoria di Lavoro')
plt.title('Stipendio Medio per Categoria di Lavoro')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest salary at the top
plt.grid(axis='x', linestyle='--', alpha=0.7)

for index, value in enumerate(average_salary_by_category['Average Salary in USD']):
    plt.text(value, index, f'{value:.0f}', va='center', ha='left', color='black')

plt.show()

```

```python
most_common_jobs = data_jobs_df['job_title'].value_counts().reset_index()
most_common_jobs.columns = ['Job Title', 'Count']

top_10_common_jobs = most_common_jobs.head(10)

top_10_common_jobs['Percentage'] = (top_10_common_jobs['Count'] / top_10_common_jobs['Count'].sum()) * 100

plt.figure(figsize=(12, 8))
plt.barh(top_10_common_jobs['Job Title'], top_10_common_jobs['Percentage'], color='#87CEEB')
plt.xlabel('Percentuale di Posizioni (%)')
plt.ylabel('Titolo di Lavoro')
plt.title('I 10 Lavori PiÃ¹ Richiesti nel Settore dei Dati (Percentuale)')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest percentage at the top
plt.grid(axis='x', linestyle='--', alpha=0.7)

for index, value in enumerate(top_10_common_jobs['Percentage']):
    plt.text(value, index, f'{value:.2f}%', va='center', ha='left', color='black')

plt.show()

```

```python
data1 = pd.read_csv('C:/Users/user/Desktop/big_tech_stock_prices.csv')

data1['date'] = pd.to_datetime(data1['date'])

selected_companies = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META']  # Apple, Microsoft, Google, Amazon, Meta (Facebook)
subset_data = data1[data1['stock_symbol'].isin(selected_companies)]

pivot_data = subset_data.pivot(index='date', columns='stock_symbol', values='close')

plt.figure(figsize=(14, 8))
for company in selected_companies:
    plt.plot(pivot_data.index, pivot_data[company], label=company)

plt.title('Prezzi di Chiusura delle Principali Aziende Tech (2010 - Presente)')
plt.xlabel('Data')
plt.ylabel('Prezzo di Chiusura')
plt.legend(title='Azienda')
plt.grid(True)
plt.show()

#https://www.kaggle.com/datasets/umerhaddii/big-tech-giants-stock-price-data
```

```python

data = pd.read_csv('C:/Users/user/Desktop/Dataset - Factors Influencing Technology Adoption in Consumer Households.csv')

import matplotlib.pyplot as plt

# Visualizzazione: Frequenza d'Uso delle Tecnologie per Gruppo di EtÃ 
age_bins = [0, 20, 30, 40, 50, 60, 70, 100]
age_labels = ['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71+']
data['GruppoEtÃ '] = pd.cut(data['Age'], bins=age_bins, labels=age_labels)

plt.figure(figsize=(12, 8))
for tech in ['Frequency_LaptopDesktop', 'Frequency_Smartphone', 'Frequency_Tablet']:
    plt.plot(data.groupby('GruppoEtÃ ')[tech].mean(), label=tech)
    
plt.title('Frequenza Media d\'Uso delle Tecnologie per Gruppo di EtÃ ')
plt.xlabel('Gruppo di EtÃ ')
plt.ylabel('Frequenza Media')
plt.legend()
plt.grid(True)
plt.show()

# Visualizzazione: Barriere all'Adozione
barriers = ['Barrier_HighPrice', 'Barrier_TechKnowledge', 'Barrier_PrivacySecurity', 'Barrier_ReliableInternet', 'Barrier_ChangeReluctance']
barrier_means = data[barriers].mean()

plt.figure(figsize=(10, 6))
barrier_means.plot(kind='bar', color='salmon', edgecolor='black')
plt.title('Livello Medio delle Barriere all\'Adozione della Tecnologia')
plt.xlabel('Barriera')
plt.ylabel('Livello Medio della Barriera')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

```

```python
# Visualizzazione: Percentuale di Uso dei Vari Apparecchi Elettronici


frequency_columns = ['Frequency_LaptopDesktop', 'Frequency_Smartphone', 'Frequency_Tablet', 'Frequency_SmartHomeDevices', 'Frequency_WearableDevices', 'Frequency_GamingConsoles']
frequency_data = data[frequency_columns]


frequency_percentages = frequency_data.apply(lambda x: x.value_counts(normalize=True) * 100)

plt.figure(figsize=(14, 8))

for column in frequency_columns:
    plt.plot(frequency_percentages.index, frequency_percentages[column], marker='o', label=column)

plt.title('Percentuale di Uso dei Vari Apparecchi Elettronici')
plt.xlabel('Frequenza di Uso')
plt.ylabel('Percentuale')
plt.legend()
plt.grid(True)
plt.show()
```

```python


percentuali_frequenze = frequency_data.apply(lambda x: x.value_counts(normalize=True) * 100).fillna(0)

percentuali_frequenze.plot(kind='bar', figsize=(14, 8), edgecolor='black')

plt.title('Percentuale di Uso dei Vari Apparecchi Elettronici')
plt.xlabel('Frequenza di Uso')
plt.ylabel('Percentuale')
plt.legend(title='Dispositivo Elettronico', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.xticks(rotation=0)
plt.show()
```

```python


plt.figure(figsize=(12, 8))
plt.boxplot([data[data['GruppoEtÃ '] == age_group]['LikelinessToRecommend'].dropna() for age_group in age_labels], labels=age_labels)
plt.title('Propensione a Raccomandare in base all\'EtÃ ')
plt.xlabel('Gruppo di EtÃ ')
plt.ylabel('Propensione a Raccomandare')
plt.grid(True)
plt.show()
```

```python
correlation_data = data[['Barrier_HighPrice', 'Barrier_TechKnowledge', 'Barrier_PrivacySecurity', 'Barrier_ReliableInternet', 'Barrier_ChangeReluctance', 'LikelinessToRecommend']].corr()

plt.figure(figsize=(10, 8))
plt.imshow(correlation_data, cmap='coolwarm', interpolation='none', aspect='auto')
plt.colorbar()
plt.xticks(range(len(correlation_data.columns)), correlation_data.columns, rotation=45)
plt.yticks(range(len(correlation_data.columns)), correlation_data.columns)
plt.title('Correlazione tra Barriere e Propensione a Raccomandare')
plt.show()
```
